# AdaBoost Classifier & Regressor with GridSearchCV Hyperparameter Tuning

This notebook demonstrates AdaBoost models using GridSearchCV for hyperparameter optimization. We compare:
- **AdaBoost Classifier** with Decision Tree base estimator
- **AdaBoost Classifier** with alternative base estimator (SVM)
- **AdaBoost Regressor** with Decision Tree base estimator
- **AdaBoost Regressor** with alternative base estimator (SVR)

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')

# Ensemble models
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor

# Base estimators
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.svm import SVC, SVR
from sklearn.linear_model import LogisticRegression, LinearRegression

# Model selection
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix, classification_report,
                             mean_squared_error, r2_score, mean_absolute_error)

# Dataset loading
from sklearn.datasets import load_breast_cancer, load_diabetes, load_wine

print("✓ All libraries imported successfully!")
```

## 2. Load and Prepare Datasets

```python
# Load Classification Dataset (Breast Cancer)
cancer = load_breast_cancer()
X_class = cancer.data
y_class = cancer.target

# Train-test split for classification
X_class_train, X_class_test, y_class_train, y_class_test = train_test_split(
    X_class, y_class, test_size=0.2, random_state=42, stratify=y_class
)

print("Classification Dataset (Breast Cancer):")
print(f"  Training set: {X_class_train.shape}")
print(f"  Test set: {X_class_test.shape}")
print(f"  Classes: {np.unique(y_class)}\n")

# Load Regression Dataset (Diabetes)
diabetes = load_diabetes()
X_reg = diabetes.data
y_reg = diabetes.target

# Train-test split for regression
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

print("Regression Dataset (Diabetes):")
print(f"  Training set: {X_reg_train.shape}")
print(f"  Test set: {X_reg_test.shape}")
print(f"  Target range: [{y_reg.min():.2f}, {y_reg.max():.2f}]")
```

## 3. AdaBoost Classifier with Decision Tree Base Estimator

```python
# Define base estimator (Decision Tree)
dt_classifier = DecisionTreeClassifier(random_state=42)

# Create AdaBoost Classifier with Decision Tree
ada_dt_classifier = AdaBoostClassifier(
    estimator=dt_classifier,
    random_state=42
)

# Define parameter grid for GridSearchCV
param_grid_dt_clf = {
    'n_estimators': [10, 25, 50, 100],
    'learning_rate': [0.5, 0.8, 1.0, 1.2],
    'estimator__max_depth': [1, 2, 3, 4],
    'estimator__min_samples_split': [2, 5]
}

print("Training AdaBoost Classifier with Decision Tree...")
print(f"Parameter grid size: {np.prod([len(v) for v in param_grid_dt_clf.values()])} combinations\n")

# GridSearchCV
grid_search_dt_clf = GridSearchCV(
    estimator=ada_dt_classifier,
    param_grid=param_grid_dt_clf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=0
)

grid_search_dt_clf.fit(X_class_train, y_class_train)

print("✓ Training completed!")
print(f"\nBest Parameters: {grid_search_dt_clf.best_params_}")
print(f"Best Cross-Validation Score: {grid_search_dt_clf.best_score_:.4f}\n")

# Predictions and evaluation
y_pred_dt_clf = grid_search_dt_clf.predict(X_class_test)

print("Classification Metrics:")
print(f"Accuracy: {accuracy_score(y_class_test, y_pred_dt_clf):.4f}")
print(f"Precision: {precision_score(y_class_test, y_pred_dt_clf):.4f}")
print(f"Recall: {recall_score(y_class_test, y_pred_dt_clf):.4f}")
print(f"F1-Score: {f1_score(y_class_test, y_pred_dt_clf):.4f}")
```

```python
# Confusion Matrix
cm_dt_clf = confusion_matrix(y_class_test, y_pred_dt_clf)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_dt_clf, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.title('Confusion Matrix: AdaBoost Classifier (Decision Tree Base)', fontsize=12, fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

## 4. AdaBoost Classifier with Alternative Base Estimator (SVM)

```python
# Define base estimator (SVM)
svm_classifier = SVC(kernel='rbf', random_state=42, probability=True)

# Create AdaBoost Classifier with SVM
ada_svm_classifier = AdaBoostClassifier(
    estimator=svm_classifier,
    random_state=42
)

# Define parameter grid for GridSearchCV
param_grid_svm_clf = {
    'n_estimators': [10, 25, 50, 75],
    'learning_rate': [0.8, 1.0, 1.2],
    'estimator__C': [0.5, 1.0, 2.0],
    'estimator__gamma': ['scale', 'auto']
}

print("Training AdaBoost Classifier with SVM...")
print(f"Parameter grid size: {np.prod([len(v) for v in param_grid_svm_clf.values()])} combinations\n")

# GridSearchCV
grid_search_svm_clf = GridSearchCV(
    estimator=ada_svm_classifier,
    param_grid=param_grid_svm_clf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=0
)

grid_search_svm_clf.fit(X_class_train, y_class_train)

print("✓ Training completed!")
print(f"\nBest Parameters: {grid_search_svm_clf.best_params_}")
print(f"Best Cross-Validation Score: {grid_search_svm_clf.best_score_:.4f}\n")

# Predictions and evaluation
y_pred_svm_clf = grid_search_svm_clf.predict(X_class_test)

print("Classification Metrics:")
print(f"Accuracy: {accuracy_score(y_class_test, y_pred_svm_clf):.4f}")
print(f"Precision: {precision_score(y_class_test, y_pred_svm_clf):.4f}")
print(f"Recall: {recall_score(y_class_test, y_pred_svm_clf):.4f}")
print(f"F1-Score: {f1_score(y_class_test, y_pred_svm_clf):.4f}")
```

```python
# Confusion Matrix
cm_svm_clf = confusion_matrix(y_class_test, y_pred_svm_clf)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_svm_clf, annot=True, fmt='d', cmap='Greens', cbar=True)
plt.title('Confusion Matrix: AdaBoost Classifier (SVM Base)', fontsize=12, fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

## 5. AdaBoost Regressor with Decision Tree Base Estimator

```python
# Define base estimator (Decision Tree)
dt_regressor = DecisionTreeRegressor(random_state=42)

# Create AdaBoost Regressor with Decision Tree
ada_dt_regressor = AdaBoostRegressor(
    estimator=dt_regressor,
    random_state=42
)

# Define parameter grid for GridSearchCV
param_grid_dt_reg = {
    'n_estimators': [10, 25, 50, 100],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'loss': ['linear', 'square', 'exponential'],
    'estimator__max_depth': [2, 3, 4, 5],
    'estimator__min_samples_split': [2, 3]
}

print("Training AdaBoost Regressor with Decision Tree...")
print(f"Parameter grid size: {np.prod([len(v) for v in param_grid_dt_reg.values()])} combinations\n")

# GridSearchCV
grid_search_dt_reg = GridSearchCV(
    estimator=ada_dt_regressor,
    param_grid=param_grid_dt_reg,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=0
)

grid_search_dt_reg.fit(X_reg_train, y_reg_train)

print("✓ Training completed!")
print(f"\nBest Parameters: {grid_search_dt_reg.best_params_}")
print(f"Best Cross-Validation Score (R²): {grid_search_dt_reg.best_score_:.4f}\n")

# Predictions and evaluation
y_pred_dt_reg = grid_search_dt_reg.predict(X_reg_test)

print("Regression Metrics:")
print(f"R² Score: {r2_score(y_reg_test, y_pred_dt_reg):.4f}")
print(f"MAE: {mean_absolute_error(y_reg_test, y_pred_dt_reg):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_reg_test, y_pred_dt_reg)):.4f}")
```

```python
# Visualization: Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_reg_test, y_pred_dt_reg, alpha=0.6, edgecolors='k')
plt.plot([y_reg_test.min(), y_reg_test.max()], [y_reg_test.min(), y_reg_test.max()], 
         'r--', lw=2, label='Perfect Prediction')
plt.xlabel('Actual Values', fontsize=11)
plt.ylabel('Predicted Values', fontsize=11)
plt.title('Actual vs Predicted: AdaBoost Regressor (Decision Tree Base)', fontsize=12, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. AdaBoost Regressor with Alternative Base Estimator (SVR)

```python
# Define base estimator (SVR)
svr_regressor = SVR(kernel='rbf')

# Create AdaBoost Regressor with SVR
ada_svr_regressor = AdaBoostRegressor(
    estimator=svr_regressor,
    random_state=42
)

# Define parameter grid for GridSearchCV
param_grid_svr_reg = {
    'n_estimators': [10, 25, 50, 75],
    'learning_rate': [0.01, 0.05, 0.1],
    'loss': ['linear', 'square'],
    'estimator__C': [0.1, 1.0, 10.0],
    'estimator__epsilon': [0.01, 0.1]
}

print("Training AdaBoost Regressor with SVR...")
print(f"Parameter grid size: {np.prod([len(v) for v in param_grid_svr_reg.values()])} combinations\n")

# GridSearchCV
grid_search_svr_reg = GridSearchCV(
    estimator=ada_svr_regressor,
    param_grid=param_grid_svr_reg,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=0
)

grid_search_svr_reg.fit(X_reg_train, y_reg_train)

print("✓ Training completed!")
print(f"\nBest Parameters: {grid_search_svr_reg.best_params_}")
print(f"Best Cross-Validation Score (R²): {grid_search_svr_reg.best_score_:.4f}\n")

# Predictions and evaluation
y_pred_svr_reg = grid_search_svr_reg.predict(X_reg_test)

print("Regression Metrics:")
print(f"R² Score: {r2_score(y_reg_test, y_pred_svr_reg):.4f}")
print(f"MAE: {mean_absolute_error(y_reg_test, y_pred_svr_reg):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_reg_test, y_pred_svr_reg)):.4f}")
```

```python
# Visualization: Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_reg_test, y_pred_svr_reg, alpha=0.6, edgecolors='k', color='orange')
plt.plot([y_reg_test.min(), y_reg_test.max()], [y_reg_test.min(), y_reg_test.max()], 
         'r--', lw=2, label='Perfect Prediction')
plt.xlabel('Actual Values', fontsize=11)
plt.ylabel('Predicted Values', fontsize=11)
plt.title('Actual vs Predicted: AdaBoost Regressor (SVR Base)', fontsize=12, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 9. Parameter Explanation & GridSearchCV Details

### AdaBoost Classifier Parameters:
- **n_estimators**: Number of weak learners (boosting rounds). More estimators = longer training time but potentially better performance.
- **learning_rate**: Shrinks the contribution of each classifier. Lower values require more estimators but may yield better results.
- **estimator**: Base learner (DecisionTreeClassifier or SVC).
  - For DecisionTree: `max_depth` controls tree depth, `min_samples_split` controls split threshold.
  - For SVC: `C` is regularization parameter, `gamma` controls kernel coefficient.

### AdaBoost Regressor Parameters:
- **n_estimators**: Number of weak learners (boosting rounds).
- **learning_rate**: Contribution weight of each regressor.
- **loss**: Loss function - 'linear', 'square', or 'exponential' (affects how errors are weighted).
- **estimator**: Base learner (DecisionTreeRegressor or SVR).
  - For DecisionTree: `max_depth`, `min_samples_split` control tree structure.
  - For SVR: `C` is regularization, `epsilon` is tolerance margin, `gamma` is kernel coefficient.

### GridSearchCV Benefits:
✓ Exhaustive search through parameter space  
✓ Automatic cross-validation for robust evaluation  
✓ Parallel processing (`n_jobs=-1` uses all CPU cores)  
✓ Easy identification of optimal hyperparameters  
✓ Returns best model automatically
